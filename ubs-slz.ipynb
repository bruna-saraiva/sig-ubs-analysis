```python
import pandas as pd
```

```python
df = pd.read_csv("CNES_UBS_Sao-luis.csv", sep=',')
```

```python
df.columns
```

```python
df.head(4)
```

```python
# Se o campo location estiver no formato:
# -2.6923404,-44.30218
df[['latitude', 'longitude']] = df['location'].str.split(',', expand=True)
# Converter para números
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)

df.head()
```

```python
df.to_csv(
    "v2-CNES_UBS_Sao-luis.csv",
    index=False,
    encoding="utf-8-sig"
)
```
